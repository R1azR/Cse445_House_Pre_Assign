```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #standardizes features 
ready_data_scaled = pd.DataFrame(scaler.fit_transform(ready_data.iloc[:, :-1]), columns=ready_data.columns[:-1])
ready_data_scaled["MEDV"] = ready_data["MEDV"]
```

```python
from sklearn.model_selection import train_test_split

#spliting train and test part with 80 to 20 
X = ready_data_scaled.drop("MEDV", axis=1)
Y = ready_data_scaled["MEDV"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42) #20% data is for testing and 80% for training
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

#model creation
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Machine": SVR()
}
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print(f"{name} - MSE: {mean_squared_error(Y_test, Y_pred):.4f}, MAE: {mean_absolute_error(Y_test, Y_pred):.4f}, R2 Score: {r2_score(Y_test, Y_pred):.4f}")
    #Calculate MSE,MAE,R2 Score for Linear Regression, Decision Tree, Random Forest, Support Vector Machine
```

```python
grid = {
    "Decision Tree": {"max_depth": [5, 10, 15, 20]},
    "Random Forest": {"n_estimators": [50, 100, 150, 200]}
    
}
```

```python
from sklearn.model_selection import GridSearchCV

for name, params in grid.items():
    search = GridSearchCV(models[name], param_grid=params, cv=5) #applying Grid Search
    search.fit(X_train, Y_train)
    print(f"Best params for {name}: {search.best_params_}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

for name, params in grid.items():
    search2 = RandomizedSearchCV(models[name], param_distributions=params, n_iter=10, cv=5, random_state=42) #applying Randomized Search
    search2.fit(X_train, Y_train)
    print(f"Best params for {name}: {search.best_params_}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

percentile = np.percentile(Y, [33, 66]) #divide continous target variable
Y_class = np.digitize(Y, percentile, right=True)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y_class, test_size=0.2, random_state=42)

classifiers = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True)
}
```

```python
from sklearn.metrics import  precision_score, recall_score, f1_score, accuracy_score

for name, model in classifiers.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print(f"{name} -  Precision: {precision_score(Y_test, Y_pred, average='weighted'):.4f}, Recall: {recall_score(Y_test, Y_pred, average='weighted'):.4f}, F1 Score: {f1_score(Y_test, Y_pred, average='weighted'):.4f}, Accuracy: {accuracy_score(Y_test, Y_pred):.4f}\n")
```

```python
from sklearn.metrics import roc_curve, auc

plot.figure(figsize=(10, 8)) #plot size
for name, model in classifiers.items():
    Y_prob = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(Y_test, Y_prob, pos_label=1) #return fpr,tpr values. ROC-> Receiver Operating Characteristic
    plot.plot(fpr, tpr, label=f"{name} (AUC = {auc(fpr, tpr):.2f}") #AUC-> Area Under the Curve
plot.xlabel("False Positive Rate")
plot.ylabel("True Positive Rate")
plot.legend()  #Show AUC values
plot.title("ROC Curve")
plot.show()
```
